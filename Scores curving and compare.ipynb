Students of a different section of the course taught by the same professor scored the following points out of 100. Write a python program to generate the quantile-quantile plot for the scores obtained by the students of the two sections. Which section performed better? Justify your answer based on statistical analysis.

20, 49, 85, 17, 33, 62, 66, 64, 37, 81, 22, 18, 45, 42, 14, 39, 67, 47, 53, 73, 58, 84, 21 

This question is elaborated on the student curving.

```python
from scipy import *
import statsmodels.graphics.gofplots
import numpy as np
import matplotlib.pyplot as pl
```

```python
student_score_section1 =  47, 63, 71, 39, 47, 49, 43, 37, 81, 69, 38, 13, 29, 61, 49, 53, 57, 23, 58, 17, 73, 33, 2
student_score_section2 =  20, 49, 85, 17, 33, 62, 66, 64, 37, 81, 22, 18, 45, 42, 14, 39, 67, 47, 53, 73, 58, 84, 21 
```

```python
score_1 = np.array([47, 63, 71, 39, 47, 49, 43, 37, 81, 69, 38, 13, 29, 61, 49, 53, 57, 23, 58, 17, 73, 33, 2])
score_2 = np.array([20, 49, 85, 17, 33, 62, 66, 64, 37, 81, 22, 18, 45, 42, 14, 39, 67, 47, 53, 73, 58, 84, 21])
```

```python
statsmodels.graphics.gofplots.qqplot_2samples(score_1, score_2, xlabel='Score for section 1', ylabel='Score for section 2', line='45')
```

### From the quantile quantile plot we can see that performance of section 1 is better because most of the data lies under 
### line of 45 degree angle.

### Therefore section 1 performs better than section 2
